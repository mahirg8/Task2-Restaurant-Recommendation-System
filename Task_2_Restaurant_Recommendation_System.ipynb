```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Data loading and preprocessing

```python
df = pd.read_csv('/content/Dataset .csv')
```

```python
df.head()
```

```python
#creating a refined dataframe
df_r = df[['Restaurant ID','Restaurant Name','Cuisines','Price range','Aggregate rating','Votes']]
df_r
```

```python
#handling missing values
df_r.isnull().sum()
```

```python
df_r = df_r.dropna()
```

```python
df_r.isna().sum()
```

```python
df_r.duplicated().sum()
```

```python
df_r['Restaurant Name'].duplicated().sum()
```

```python
df_r['Restaurant Name'].value_counts()
```

```python
#sorting restaurant by name and rating
df_r = df_r.sort_values(by=['Restaurant Name','Aggregate rating'],ascending=False)
df_r.head()
```

```python
df_r[df_r["Restaurant Name"]=="Cafe Coffee Day"].head()
```

```python
#removing duplicate entries of same restaurant name
df_r = df_r.drop_duplicates('Restaurant Name', keep='first')
df_r
```

```python
df_r['Restaurant Name'].value_counts()
```

```python
df_r = df_r[df_r['Aggregate rating']>3.9]
df_r
```

```python
#splitting cuisines into list
df_r['Cuisines'] = df_r['Cuisines'].str.split(',')
df_r
```

```python
df_r = df_r.explode('Cuisines')
df_r
```

```python
df_r['Cuisines'].value_counts()
```

```python
resto_cuisines = pd.crosstab(df_r['Restaurant Name'], df_r['Cuisines'])
resto_cuisines
```

```python
df_r['Restaurant Name'].sample(20, random_state=194)
```

```python
from sklearn.metrics import jaccard_score
```

```python
print(jaccard_score(resto_cuisines.loc["Olive Bistro"].values, resto_cuisines.loc["Rose Cafe"].values))
```

```python
from scipy.spatial.distance import pdist, squareform
```

```python
jaccardDist = pdist(resto_cuisines.values, metric='jaccard')
jaccardMatrix = squareform(jaccardDist)
jaccardSim = 1 - jaccardMatrix
dfJaccard = pd.DataFrame(jaccardSim, index=resto_cuisines.index, columns=resto_cuisines.index)
dfJaccard
```

```python
df_r['Restaurant Name'].sample(20)
```

**Final Recommendation System**

(use algos like RF for user input recommendation.. here it is just static rigid output)

```python
resto = "Ooma"

sim = dfJaccard.loc[resto].sort_values(ascending=False)
sim = pd.DataFrame({'Restaurant Name' : sim.index, 'simScore' : sim.values})
sim = sim[(sim['Restaurant Name'] != resto) & (sim['simScore'] > 0.7)].head(5)
RestoRec = pd.merge(sim, df_r[['Restaurant Name','Aggregate rating']], how='inner', on='Restaurant Name')
FinalRestoRec = RestoRec.sort_values('Aggregate rating', ascending=False).drop_duplicates('Restaurant Name', keep='first')
FinalRestoRec
```
